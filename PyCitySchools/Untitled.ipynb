```python
# Dependencies and Setup
import pandas as pd
```

```python
schools_complete = "Resources/schools_short.csv"
students_complete = "Resources/students_short.csv"
```

```python
schools_complete_df = pd.read_csv(schools_complete)
students_complete_df = pd.read_csv(students_complete)
schools_complete_df
```

```python
merged_school_data_df = pd.merge(schools_complete_df, students_complete_df, on="school_name")
merged_school_data_df
```

```python
bins = [0, 49.9, 100]
group_names = ["Fail", "Pass"]
pd.cut(merged_school_data_df["maths_score"], bins, labels=group_names).head()
# # pd.cut(grouped_data["maths_score"],bins, labels=group_labels).head()
merged_school_data_df["maths_score2"] = pd.cut(merged_school_data_df["maths_score"], bins, labels=group_names)
# # merged_school_data_df["budget"].sum()
```
